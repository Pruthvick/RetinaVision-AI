```python
import gradio as gr
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import load_model
from PIL import Image
```

```python
MODEL_PATH = r"D:\multidieseases\EfficientNetV2_BCE.keras"

model = load_model(MODEL_PATH)
```

```python
selected_diseases = [

    "DR",
    "ARMD",
    "MH",
    "DN",
    "MYA",
    "BRVO",
    "TSLN",
    "LS",
    "CSR",
    "ODC",
    "ODP",
    "ODE",
    "RS",
    "CRS"

]
```

```python
def predict_retinal_disease(image):

    # Resize image
    image = image.resize((224,224))

    # Convert to numpy
    image = np.array(image).astype(np.float32)

    # RGB check
    if image.shape[-1] == 4:
        image = image[:,:,:3]

    # Batch dimension
    image = np.expand_dims(image, axis=0)

    # Predict
    predictions = model.predict(image, verbose=0)[0]

    threshold = 0.5

    result = {}

    detected = False

    for disease, prob in zip(selected_diseases, predictions):

        if prob >= threshold:

            result[disease] = float(prob)

            detected = True

    if not detected:

        result["Normal Retina"] = 1.0

    return result
```

```python
demo = gr.Interface(

    fn=predict_retinal_disease,

    inputs=gr.Image(type="pil"),

    outputs=gr.Label(num_top_classes=14),

    title="Retinal Disease Detection",

    description="Upload a Fundus Image"

)
```

```python
demo.launch()
```
